Библиотека Simpy.
SymPy — это библиотека Python для символьной математики. Она стремится стать полнофункциональной системой компьютерной алгебры (CAS), сохраняя при этом код максимально простым, чтобы быть понятным и легко расширяемым. Она может выполнять от обычной базовой арифметики, до гамма-функций , дзета-функций , полином. Для понятности, тут будет написано как раз про базовую арифметику, уравнения.

Начнем с корней:
калькулятор не выводит иррациональное число, и заменяет его на красивое алгебраическое выражение

```python
import sympy
print(sympy.sqrt(3))
print(sympy.sqrt(4))
print(sympy.sqrt(8))
```

дальше, SymPy способнен вычислять символьные выражения с переменными.

Как мы увидим позже, в SymPy переменные определяются с помощью symbols. В отличие от многих систем символьной манипуляции, переменные в SymPy должны быть определены до их использования.

Определим символическое выражение, представляющее математическое выражение  x + 2y:

```python
from sympy import symbols
x, y = symbols('x y')
expr = x + 2*y
expr
```

дальше, будут выражения по сложнее из тригонометрии, SimPy способен решить их всех. Для этого нужно ввести import sympy import *, ( * - cos, integrate, и тд.)

```python
import sympy
from sympy import *
diff(sin(x)*exp(x), x)

```

```python
from sympy import *
integrate(exp(x)*sin(x) + exp(x)*cos(x), x)
```

```python
from sympy import *
integrate(sin(x**2), (x, -oo, oo))
```

квадратное уравнение:

```python
from sympy import *
solve(x**2 - 2, x)
```

эта библиотека может решить всё.

```python
from sympy import *
expr = x**2 + x*y
srepr(expr)
Add(Pow(Symbol('x'), Integer(2)), Mul(Symbol('x'), Symbol('y')))
```
